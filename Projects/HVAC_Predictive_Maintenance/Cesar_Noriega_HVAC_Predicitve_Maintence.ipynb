```python
import pandas as pd
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
```

```python
rows = []
start_time = datetime.now()

for i in range(120):
    data = {
        "timestamp": start_time + timedelta(minutes=i*30),
        "temperature": random.randint(80, 110),
        "vibration": round(random.uniform(1.0, 5.0), 2),
        "humidity": round(random.uniform(50, 90), 2)
    }

    if data["temperature"] > 95 and data["vibration"] > 3:
        data["failure"] = 1
    else:
        data["failure"] = 0

    rows.append(data)

df = pd.DataFrame(rows)
df.head()
```

```python
plt.figure()
plt.plot(df["timestamp"], df["temperature"])
plt.title("HVAC Temperature Over Time")
plt.xlabel("Time")
plt.ylabel("Temperature (°F)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
failures = df["failure"].sum()

print("===== SYSTEM RESULT =====")

if failures > 15:
    print("Status: WARNING")
else:
    print("Status: NORMAL")
```
